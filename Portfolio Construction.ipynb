```python
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr
import yfinance as yf
from pylab import plt, mpl
plt.style.use("seaborn")
mpl.rcParams['font.family'] = "serif"

%matplotlib inline
yf.pdr_override()
```

```python
tickers = ["AAPL", "MSFT", "META", "NVDA", "TSLA", "AMZN", "SMCI", "JPM", "KO", "MCD"]
df = pdr.get_data_yahoo(tickers, start = "1990-01-01", end = "2024-11-10")['Adj Close'].dropna()
df
```

```python
import os 
dirs = os.listdir()
files = [dir for dir in dirs if ("ASTUTE" in dir)]
names = [dir.replace(".csv", "") for dir in dirs if ("ASTUTE" in dir)]
dfs = [pd.read_csv(file, index_col=1)[['NAV Price']].rename(columns={"NAV Price": file.replace(".csv", "")}) for file in files]
df = pd.DataFrame()

for temp_df in dfs:
     df = pd.concat([df, temp_df], axis=1).dropna()
df
```

```python
rets = (np.log(df / df.shift(1))).dropna()
rets.hist(bins=40, figsize=(10,8));
```

```python
total_return = (1 + rets).prod() - 1
annualized_return = ((total_return + 1) ** (252/rets.shape[0]) ) - 1

weightages = np.abs(np.random.standard_normal(len(rets.columns)))
weightages /= weightages.sum()

risk_free_rate = 0.05
```

## Portfolio Return

```python
portfolio_ret = annualized_return @ weightages
```

## Portfolio Risk

```python
np.sqrt(weightages @ rets.cov() * 252 @ weightages.T)
```

```python
def portfolio_ret(weights):
    return annualized_return @ weights
def portfolio_risk(weights):
    return np.sqrt(weights @ rets.cov() * 252 @ weights.T)
```

```python
eweights = []
port_rets = []
port_risks = []
for i in range(10000):
    weightages = np.abs(np.random.standard_normal(len(rets.columns)))
    weightages /= weightages.sum()
    eweights.append(weightages)
    port_rets.append(portfolio_ret(weightages))
    port_risks.append(portfolio_risk(weightages))

port_rets = np.array(port_rets)
port_risks = np.array(port_risks)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(port_risks, port_rets, c=(port_rets - risk_free_rate) / port_risks, cmap="coolwarm");
plt.colorbar(label="Sharpe Ratio");
```

## Determine the Min Var Portoflio

```python
import scipy.optimize as sco
```

```python
## objective = minimize portfolio_risk
constraint = ({"type": "eq", "fun": lambda x: np.sum(x) - 1})
eweights = np.array(len(rets.columns) * [1. / len(rets.columns)])
bns = tuple((0,1) for i in range(len(eweights)))
min_var_func = sco.minimize(portfolio_risk, eweights, constraints=constraint, bounds=bns, method="SLSQP")
min_var_func
```

```python
portfolio_risk(min_var_func["x"])
```

## Determine Optimal Portfolio

```python
def min_sharpe_ratio(weights):
    return -1 * ((portfolio_ret(weights) - risk_free_rate) / portfolio_risk(weights))
```

```python
tangent_porfolio = sco.minimize(min_sharpe_ratio, eweights, method="SLSQP", bounds=bns, constraints=constraint)
tangent_porfolio
```

## Efficient Frontier

```python
min_rets = 0.2
max_rets = port_rets.max()
ef_x = np.linspace(min_rets, max_rets, 50)
constraint = ({"type": "eq", "fun": lambda x: np.sum(x) - 1},
             {"type": "eq", "fun": lambda x: portfolio_ret(x) - target_return })
t_rets = []; t_risks = []
for target_return in ef_x:
    efficient_frontier = sco.minimize(portfolio_risk, eweights, method='SLSQP', constraints=constraint, bounds=bns)
    t_rets.append(portfolio_ret(efficient_frontier['x']))
    t_risks.append( efficient_frontier['fun'])
t_rets = np.array(t_rets)
t_risks = np.array(t_risks)
```

## Determine Capital Market Line

```python
x2 = portfolio_risk(tangent_porfolio['x'])
y2 = portfolio_ret(tangent_porfolio['x'])
b = (y2 - risk_free_rate) / x2
b
```
